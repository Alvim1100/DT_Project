```python
import os
import spotipy
import pandas as pd
import pylast
from spotipy.oauth2 import SpotifyOAuth
from dotenv import load_dotenv

load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope="user-library-read"))
```

```python
DT_uri = '2aaLAng2L2aWD2FClzwiep'

results = sp.artist_albums(DT_uri, album_type='album', limit=10)
albums = results['items']
while results['next']:
    results = sp.next(results)
    albums.extend(results['items'])
```

```python
rows = []
for album in albums:
    musics = sp.album_tracks(album['id'])['items']
    musics_ids = [m['id'] for m in musics]

    for music in musics:
        row = {
            'album_name' : album['name'],
            'release_date' : album['release_date'],
            'music_name' : music['name'],
            'music_id' : music['id'],
            'duration_ms' : music['duration_ms'],
        }
        rows.append(row)

df = pd.DataFrame(rows)
```

```python
timedeltas = pd.to_timedelta(df['duration_ms'], unit = 'ms')

df['duration_min'] = df['duration_ms'] / 60000
df['duration_formatted'] = timedeltas.apply(lambda x: f"{x.components.minutes + (x.components.hours*60):02d}:{x.components.seconds:02d}")
```

```python
remove_words = 'Live|Archives|Special|Lost Not Forgotten|Demos|2007'

df_clean = df[~df['album_name'].str.contains(remove_words, case=False, na=False)]
df = df_clean
df.reset_index(drop=True, inplace=True)
df.set_index(['album_name', 'music_name'], inplace=True)
```
